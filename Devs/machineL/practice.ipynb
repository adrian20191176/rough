## Data Cleaning Practise

```python
details = pd.read_csv('details.csv')
```

```python
details.head()
```

```python
details.columns
```

```python
pp.plot(details.loc[0,'1 star':'5 star'])
```

```python
bshort = details[['appId','app Title','Downlaods','Rounded Rating','1 star','2 star','3 star','4 star','5 star','size', 'android Version', 'appDeveloper',
       'Summary', 'Description']]
```

```python
bshort.head()
```

```python
bshort.columns = ['appId','app Title','Downlaods','Rounded Rating','1 star','2 star','3 star','4 star','5 star','size', 'android Version', 'appDeveloper',
       'Summary', 'Description']
bshort.head()
```

```python
bmelt = bshort.melt(['appId','app Title','Downlaods','Rounded Rating','android Version', 'appDeveloper',
       'Summary', 'Description'],['1 star','2 star','3 star','4 star','5 star'],'StarRating','size')
bmelt.head()
```

```python
bmelt.query('appDeveloper == "Strava+Inc."')
```

```python
bfinal = bmelt[['appId','app Title','Downlaods','Rounded Rating','android Version', 'appDeveloper',
       'Summary', 'Description','StarRating','size']]
bfinal.sort_values(['appId','appDeveloper'],inplace=True)
```

bfinal.head()

```python
bfinal.head()
```

```python
tracks = bfinal[['appId','app Title','Downlaods','Rounded Rating','android Version', 'appDeveloper']].drop_duplicates()
tracks.head()
```

```python
tracks.index.name = 'Id'
tracksappId = tracks.reset_index()
tracksappId.head()
```

```python
pd.merge(tracksappId,bfinal,on=['appId','appDeveloper']).head()
```

```python
pd.merge(tracksappId,bfinal,on=['appId','appDeveloper','Downlaods']).head()
```

```python
tidy = pd.merge(tracksappId,bfinal,on=['appId','appDeveloper','Downlaods']).drop(['appId','appDeveloper','Downlaods'],axis=1)
tidy.head()
```
